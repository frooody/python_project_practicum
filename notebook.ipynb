## Описание проекта

Данный проект посвящен анализу лояльности пользователей сервиса Яндекс Афиша. Основная цель исследования — выявить факторы, влияющие на возврат пользователей на платформу для совершения повторных покупок билетов.
## Исходные данные

Выгрузка из базы данных SQL должна позволить собрать следующие данные:

- `user_id` — уникальный идентификатор пользователя, совершившего заказ;
- `device_type_canonical` — тип устройства, с которого был оформлен заказ (`mobile` — мобильные устройства, `desktop` — стационарные);
- `order_id` — уникальный идентификатор заказа;
- `order_dt` — дата создания заказа (используйте данные `created_dt_msk`);
- `order_ts` — дата и время создания заказа (используйте данные `created_ts_msk`);
- `currency_code` — валюта оплаты;
- `revenue` — выручка от заказа;
- `tickets_count` — количество купленных билетов;
- `days_since_prev` — количество дней от предыдущей покупки пользователя, для пользователей с одной покупкой — значение пропущено;
- `event_id` — уникальный идентификатор мероприятия;
- `service_name` — название билетного оператора;
- `event_type_main` — основной тип мероприятия (театральная постановка, концерт и так далее);
- `region_name` — название региона, в котором прошло мероприятие;
- `city_name` — название города, в котором прошло мероприятие.

````python
dotenv.load_dotenv()
connection_string = 'postgresql://{}:{}@{}:{}/{}'.format(
    os.getenv('DB_USER'),
    os.getenv('DB_PASSWORD'),
    os.getenv('DB_HOST'),
    os.getenv('DB_PORT'),
    os.getenv('DB_NAME'),
) 
engine = create_engine(connection_string) 
````

````python
query = '''
    -- Настройка параметра synchronize_seqscans важна для проверки
    WITH set_config_precode AS (
    SELECT set_config('synchronize_seqscans', 'off', true)
    )
    SELECT
    s.user_id,
    device_type_canonical,
    order_id ,
    created_dt_msk order_dt,
    created_ts_msk order_ts,
    currency_code ,
    revenue ,
    tickets_count ,
    (created_dt_msk)::date - (LAG(created_dt_msk) OVER (PARTITION BY user_id ORDER BY created_dt_msk))::date AS days_since_prev,
    s.event_id,
    e.event_name_code as event_name,
    e.event_type_main ,
    service_name ,
    r.region_name ,
    c.city_name
    from afisha.purchases s
    LEFT JOIN afisha.events e ON s.event_id = e.event_id 
    LEFT JOIN afisha.city c  ON e.city_id = c.city_id 
    LEFT JOIN afisha.regions r ON r.region_id  = c.region_id 
    WHERE device_type_canonical IN ('mobile', 'desktop') and event_type_main != 'фильм'
    ORDER BY user_id 
''' 
df = pd.read_sql_query(query, con=engine) 
df.head()
````

````python
def segment_summary(df, column, title, top_n=None):
    seg = df.groupby(column).agg(
        users_count=('user_id', 'count')
    ).reset_index().sort_values('users_count', ascending=False)
    
    seg['users_share'] = (seg['users_count'] / seg['users_count'].sum() * 100).round(2)
    
    # Ограничиваем данные ДО построения графика
    if top_n and len(seg) > top_n:
        seg_plot = seg.head(top_n)
    else:
        seg_plot = seg
    
    plt.figure(figsize=(10, 6))
    seg_plot.plot(kind='bar', x=column, y='users_share')
    plt.title(title)
    plt.xlabel(column)
    plt.ylabel('Доля пользователей (%)')
    plt.xticks(rotation=45, ha='right')
    plt.show()
    
    return seg_plot

event_dist = segment_summary(user_profile, 'first_event_type', 'Распределение по типу мероприятия')
device_dist = segment_summary(user_profile, 'first_device', 'Распределение по типу устройства')
region_dist = segment_summary(user_profile, 'first_region', 'Распределение по регионам (топ-10)', top_n=10)
partner_dist = segment_summary(user_profile, 'first_partner', 'Распределение по партнёрам (топ-10)', top_n=10)

print('Общая картина распределения пользователей:')
print(f'Всего пользователей: {len(user_profile)}')
print(f'\nТоп-3 типа мероприятия:')
print(event_dist.head(3)[['first_event_type', 'users_count', 'users_share']])
print(f'\nТоп-3 региона:')
print(region_dist.head(3)[['first_region', 'users_count', 'users_share']])
````

````python
def plot_retention_by_feature(feature, title, max_segments=10):
    retention_data = user_profile.groupby(feature).agg(
        total_users=('user_id', 'count'),
        returned_users=('is_two', 'sum')
    ).reset_index()
    retention_data['retention_rate'] = retention_data['returned_users'] / retention_data['total_users']
    
    # Берем топ-N по количеству пользователей (без условия)
    retention_data = retention_data.nlargest(max_segments, 'total_users')
    
    plt.figure(figsize=(12, 6))
    retention_data.plot(kind='bar', x=feature, y='retention_rate')
    plt.axhline(user_profile['is_two'].mean(), color='red', linestyle='--', label='Среднее по выборке')
    plt.title(title)
    plt.xlabel(feature)
    plt.ylabel('Доля вернувшихся пользователей')
    plt.xticks(rotation=45, ha='right')
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    return retention_data

result_event = plot_retention_by_feature('first_event_type', 'Возврат пользователей по типу первого мероприятия')
result_device = plot_retention_by_feature('first_device', 'Возврат пользователей по типу устройства')
result_region = plot_retention_by_feature('first_region', 'Возврат пользователей по региону (топ-10)')
result_partner = plot_retention_by_feature('first_partner', 'Возврат пользователей по партнёру (топ-10)')

overall_retention = user_profile['is_two'].mean()
print(f'Общая доля вернувшихся пользователей: {overall_retention:.2%}')

print('\nТоп-3 сегмента по возврату (event_type):')
print(result_event.sort_values('retention_rate', ascending=False).head(3)[['first_event_type', 'total_users', 'retention_rate']])
print('\nТоп-3 сегмента по возврату (device):')
print(result_device.sort_values('retention_rate', ascending=False).head(3)[['first_device', 'total_users', 'retention_rate']])
print('\nТоп-3 сегмента по возврату (region):')
print(result_region.sort_values('retention_rate', ascending=False)[['first_region', 'total_users', 'retention_rate']])
print('\nТоп-3 сегмента по возврату (partner):')
print(result_partner.sort_values('retention_rate', ascending=False).head(3)[['first_partner', 'total_users', 'retention_rate']])
````

- Постройте сравнительные гистограммы распределения средней выручки с билета (`avg_revenue_rub`):
    - для пользователей, совершивших один заказ;
    - для вернувшихся пользователей, совершивших 2 и более заказа.
- Ответьте на вопросы:
    - В каких диапазонах средней выручки концентрируются пользователи из каждой группы?
    - Есть ли различия между группами?

**Рекомендация:**

1. Используйте одинаковые интервалы (`bins`) и прозрачность (`alpha`), чтобы визуально сопоставить распределения.
2. Задайте параметру `density` значение `True`, чтобы сравнивать форму распределений, даже если число пользователей в группах отличается.

---


````python
one_order = user_profile[user_profile['total_orders'] == 1]['avg_revenue_rub']
two_plus = user_profile[user_profile['total_orders'] >= 2]['avg_revenue_rub']

plt.figure(figsize=(12, 6))
plt.hist(one_order, bins=50, alpha=0.5, label='1 заказ', density=True, edgecolor='black')
plt.hist(two_plus, bins=50, alpha=0.5, label='2+ заказа', density=True, edgecolor='black')
plt.title('Распределение средней выручки с заказа')
plt.xlabel('Средняя выручка (руб)')
plt.ylabel('Плотность')
plt.legend()
plt.show()

print('Статистика для пользователей с 1 заказом:')
print(one_order.describe())
print('\nСтатистика для пользователей с 2+ заказами:')
print(two_plus.describe())

print(f'\nСредняя выручка (1 заказ): {one_order.mean():.2f} руб., медиана: {one_order.median():.2f} руб.')
print(f'Средняя выручка (2+ заказа): {two_plus.mean():.2f} руб., медиана: {two_plus.median():.2f} руб.')
print(f'Разница средних: {two_plus.mean() - one_order.mean():.2f} руб.')

ranges = [(0, 200), (200, 400), (400, 600), (600, 800), (800, 1200), (1200, float('inf'))]
range_labels = ['0–200', '200–400', '400–600', '600–800', '800–1200', '1200+']

print('\nДоля пользователей по диапазонам выручки:')
for (low, high), label in zip(ranges, range_labels):
    one_pct = ((one_order >= low) & (one_order < high)).mean() * 100
    two_pct = ((two_plus >= low) & (two_plus < high)).mean() * 100
    print(f'{label:8} руб.: 1 заказ — {one_pct:5.1f}%, 2+ заказа — {two_pct:5.1f}%')
````

Средние значения практически одинаковы (~545–548 руб.), но медианы различаются: 378 руб. у однократных против 494 руб. у вернувшихся. Это говорит о том, что возвращающиеся пользователи в «типичном» случае тратят больше.

Ключевые различия по диапазонам:
- В диапазоне 0–200 руб. концентрируется 28% пользователей с 1 заказом против 18% вернувшихся — однократные чаще делают мелкие покупки.
- В диапазоне 400–800 руб. (наиболее массовый) доля вернувшихся выше: 45% против 38% — это «ядро» лояльной аудитории.
- В диапазоне 1200+ руб. однократных больше (15% против 9%) — крупные разовые заказы реже приводят к возврату.

Возможные причины:
- Нулевая/низкая выручка у части однократных пользователей, вероятно, связана с отменами заказов или возвратами билетов.
- Крупные заказы (1200+ руб.) у однократных — возможно, разовые групповые покупки (корпоратив, день рождения), не предполагающие повторного визита.

Такое возможно, когда в данных преобладает одно значение: в таком случае корреляционный анализ может показать отсутствие связей. Чтобы этого избежать, выделите сегменты пользователей по полю `total_orders`, а затем повторите корреляционный анализ. Выделите такие сегменты:
    - 1 заказ;
    - от 2 до 4 заказов;
    - от 5 и выше.
- Визуализируйте результат корреляции с помощью тепловой карты.

---

````python
cols_for_corr = ['total_orders', 'first_device', 'first_region', 'first_event_type', 
                 'first_partner', 'avg_revenue_rub', 'avg_tickets_count', 'avg_days_between']

correlation_matrix = phik.phik_matrix(
    user_profile[cols_for_corr], 
    interval_cols=['total_orders', 'avg_revenue_rub', 'avg_tickets_count', 'avg_days_between']
)
plt.figure(figsize=(8, 3))
sns.heatmap(correlation_matrix['total_orders'].drop('total_orders').sort_values(ascending=False).to_frame().T, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Матрица корреляций Phi-k')
plt.tight_layout()
plt.show()

print('Корреляция признаков с total_orders:')
print(correlation_matrix['total_orders'].sort_values(ascending=False))
````
